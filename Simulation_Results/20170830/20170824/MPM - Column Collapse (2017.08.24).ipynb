# Column Collapse with Rigid Boundary - Artificial Damping

## Artificial Damping

Dyanmic Equation of Motion
$$m\ddot{u} + c\dot{u} + ku = f^{ext} + f^{int} $$

Current Implementation
$$m\ddot{u} = f^{ext} + f^{int}$$

Artificial Damping ($\xi$ is damping ratio)
$$m\ddot{u} = f^{ext} + f^{int} + f^{damp}$$
$$f^{damp} = -\xi \frac{\dot{u}}{\|\dot{u}\|} \|f^{ext} - f^{int}\|$$ 

## Column Collapse (No Movement)
- Linear Elastic Model
- No initial stress
- $E = 1E7 Pa = 10 MPa$
- $\nu = 0.0$
- $dt = 1E-6 s$
- $\gamma = 2038.74 kg/m^3$
- 5% damping is used
- Spacing of Particles is $0.005$
- Spacing of Mesh is $0.01$
- Points per Cell $\approx 4$
- Quadrilateral Mesh

```python
# Import Matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import Data

nodamp_11 = np.genfromtxt("NoDamping/stress_11.txt")
nodamp_210 = np.genfromtxt("NoDamping/stress_210.txt")
nodamp_410 = np.genfromtxt("NoDamping/stress_410.txt")
nodamp_610 = np.genfromtxt("NoDamping/stress_610.txt")
nodamp_790 = np.genfromtxt("NoDamping/stress_790.txt")

damp_11 = np.genfromtxt("Damping/stress_11.txt")
damp_210 = np.genfromtxt("Damping/stress_210.txt")
damp_410 = np.genfromtxt("Damping/stress_410.txt")
damp_610 = np.genfromtxt("Damping/stress_610.txt")
damp_790 = np.genfromtxt("Damping/stress_790.txt")
```

```python
# Plotting the case with no damping against with damping
time_step = nodamp_11[:,0]

# Plot 1
plt.figure(figsize=(15,5))
plt.plot(time_step, nodamp_11[:,2], 'r', label='No Damping')
plt.plot(time_step, damp_11[:,2], 'b', label='Damping (5%)')
plt.legend(loc='lower right')
plt.title('Column Collapse (Point 11)')
plt.xlabel('Time (1E-6 s)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ylabel('Stress (Pa)')
plt.grid(True)
plt.show()

# Plot 2
plt.figure(figsize=(15,5))
plt.plot(time_step, nodamp_210[:,2], 'r', label='No Damping')
plt.plot(time_step, damp_210[:,2], 'b', label='Damping (5%)')
plt.legend(loc='lower right')
plt.title('Column Collapse (Point 210)')
plt.xlabel('Time (1E-6 s)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ylabel('Stress (Pa)')
plt.grid(True)
plt.show()

# Plot 3
plt.figure(figsize=(15,5))
plt.plot(time_step, nodamp_410[:,2], 'r', label='No Damping')
plt.plot(time_step, damp_410[:,2], 'b', label='Damping (5%)')
plt.legend(loc='lower right')
plt.title('Column Collapse (Point 410)')
plt.xlabel('Time (1E-6 s)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ylabel('Stress (Pa)')
plt.grid(True)
plt.show()

# Plot 4
plt.figure(figsize=(15,5))
plt.plot(time_step, nodamp_610[:,2], 'r', label='No Damping')
plt.plot(time_step, damp_610[:,2], 'b', label='Damping (5%)')
plt.legend(loc='lower right')
plt.title('Column Collapse (Point 610)')
plt.xlabel('Time (1E-6 s)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ylabel('Stress (Pa)')
plt.grid(True)
plt.show()

# Plot 5
plt.figure(figsize=(15,5))
plt.plot(time_step, nodamp_790[:,2], 'r', label='No Damping')
plt.plot(time_step, damp_790[:,2], 'b', label='Damping (5%)')
plt.legend(loc='lower right')
plt.title('Column Collapse (Point 790)')
plt.xlabel('Time (1E-6 s)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ylabel('Stress (Pa)')
plt.grid(True)
plt.show()
```
